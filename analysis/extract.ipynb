```python
import requests
import time

host = 'kw-extract-server'
port = 8000
endpoint = 'keywords'
URL = f'http://{host}:{port}/{endpoint}'

text = r"""
Black holes are among the most fascinating and enigmatic objects in the universe. They are regions of space where gravity is so intense that nothing, not even light, can escape once it passes the event horizon - the black hole's boundary. Formed from the remnants of massive stars that have collapsed under their own gravity, or at the centers of galaxies, black holes continue to puzzle and intrigue scientists. Their extreme nature challenges our understanding of physics, particularly at the intersection of general relativity and quantum mechanics. Despite their name, black holes are not truly empty voids. They possess enormous mass concentrated into an infinitesimally small point called a singularity. As matter and light fall into a black hole, they form an accretion disk, often emitting powerful jets of energy and particles. Black holes also distort spacetime around them, leading to phenomena like gravitational lensing and time dilation. Recent breakthroughs, such as the first image of a black hole's shadow and the detection of gravitational waves from merging black holes, have opened new avenues for studying these cosmic monsters. As our understanding grows, black holes continue to shed light on the fundamental nature of our universe.
"""

data = {
    'text': text
}

start = time.perf_counter()
response = requests.post(
    URL,
    json=data,
    headers={"Content-Type": "application/json"},
)

print(response.json())
end = time.perf_counter()
```

```python
dt = end - start
f'Time taken: {dt:.2f} s/req'
```

```python
CHUNK_SIZE = 10000
dt = dt * CHUNK_SIZE
dt /= 60 # Convert to m/req
print(f'Time for 1 chunk: {dt:.2f} m/chunk')
```
